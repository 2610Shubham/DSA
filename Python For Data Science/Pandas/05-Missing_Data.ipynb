# **`.isnull()`**

The `.isnull()` method detects missing values in a Series or DataFrame. It returns a boolean object where `True` indicates the presence of a missing value (`NaN`).




```python
df.isnull()
```

```python
df.isnull().sum()
```

---

# **`.notnull()`**

The `.notnull()` method detects non-missing values in a Series or DataFrame. It returns a boolean object where `True` indicates the presence of a valid (non-`NaN`) value.

```python
df.notnull()
```

### `Select a data where the pre-movie-score is not null`

```python
df[df['pre_movie_score'].notnull()]
```

```python
df[df['pre_movie_score'].isnull()]
```

# **`.dropna()`**

The `.dropna()` method removes missing values (`NaN`) from a DataFrame or Series. It provides options to drop rows, columns, or specific values based on the presence of `NaN`.


```python
df.dropna()
```

```python
# Specifies the minimum number of non-NaN values required to keep a row or column.
df.dropna(thresh=1) # Drops rows that do not have at least one non-NaN value.
```

```python
df.fillna(df.mean()) # It's a Dirty method to deal with the missing data 
```

```python
df['pre_movie_score'].fillna( df['pre_movie_score'].mean() )
```
