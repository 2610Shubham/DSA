# **`.sort_values()` in Pandas**

The `.sort_values()` method in Pandas is used to sort a DataFrame or Series based on the values of one or more columns. It allows sorting in ascending or descending order and provides options for handling missing data and customizing sorting behavior.


```python
df.sort_values('tip') # Sort the df based on the tip column
```

```python
df.sort_values('tip', ascending=False)
```

# **`.corr()`**

The `.corr()` method in Pandas computes pairwise correlation coefficients between numerical columns in a DataFrame. It measures the strength and direction of linear relationships between variables, with values ranging from `-1` (strong negative correlation) to `1` (strong positive correlation). A value of `0` indicates no linear correlation.


```python
df.corr()
```

# **`.value_counts()`**

The `.value_counts()` method is used to count the occurrences of unique values in a Series. It returns a Series where the index represents the unique values, and the corresponding values are their respective counts, sorted in descending order by default.


```python
df['sex'].value_counts()
```

```python
df['smoker'].value_counts()
```

```python
df['day'].value_counts()
```

# **`.unique()`**

The `.unique()` method returns the unique values in a Series or the unique values within each column of a DataFrame. It outputs a NumPy array of distinct elements, preserving the order of their appearance.


```python
df['day'].unique()
```

# **`.nunique()`**

The `.nunique()` method returns the number of unique values in a Series or, for a DataFrame, the number of unique values in each column. It can optionally include or exclude `NaN` values in the count.


```python
df['day'].nunique()
```

# **`.replace()`**

The `.replace()` method is used to replace specified values in a Series or DataFrame with new values. It supports replacing single values, lists, dictionaries, or patterns using regular expressions, providing flexibility for data cleaning and transformation tasks.


```python
df['sex'].replace(to_replace='Female', value='F')
```

```python
df['sex'].replace( ['Female' , 'Male'], ['F','M'] )
```

# **`.map()`**

The `.map()` method is used to transform each element in a Pandas Series by applying a function, dictionary, or Series. It is commonly used for element-wise operations, value replacements, or applying custom logic to modify the data.


```python
mymap = {'Female': 'F', 
         'Male': 'M'
        }
```

```python
df['sex'].map(mymap)
```

# **`.duplicated()`**

The `.duplicated()` method identifies duplicate rows in a DataFrame or duplicate elements in a Series. It returns a boolean Series indicating whether each row or element is a duplicate, with options to specify how duplicates are considered (e.g., first occurrence, last occurrence).


```python
df.duplicated()
```

```python
# Creating a sample data set to check it 
sample_df = pd.DataFrame([1,2,3,3,3],['a','b','c','d','e'])
```

```python
sample_df
```

```python
sample_df.duplicated()
```

# **`drop_duplicates()`**

The `drop_duplicates()` method removes duplicate rows from a DataFrame or duplicate elements from a Series. It provides options to specify which columns to consider for duplicates and whether to keep the first or last occurrence of each duplicate, or to drop all duplicates entirely.


```python
sample_df.drop_duplicates(inplace=True)
```

```python
sample_df
```

# **`.between()`**

The `.between()` method checks whether each value in a Series lies within a specified range. It returns a boolean Series where `True` indicates the value is within the range, inclusive of the boundary values by default.


```python
df['total_bill'].between(10,20, inclusive=True)
```

```python
df[df['total_bill'].between(10, 20, inclusive=True)]
```

# **`.nlargest()`**

The `.nlargest()` method returns the specified number of rows with the largest values in a Series or DataFrame column. It is commonly used for quickly identifying top values based on a specific criterion.


```python
df.nlargest(10, 'tip')
```

# **`.nsmallest()`**

The `.nsmallest()` method returns the first `n` smallest values from a DataFrame or Series. It sorts the data in ascending order and returns the corresponding rows or elements.


```python
df.nsmallest(10, 'tip')
```

# **`.sample()`**

The `.sample()` method randomly selects a specified number of rows or elements from a DataFrame or Series. It can be used to create random samples for data analysis, with options to set a random seed for reproducibility.


```python
df.sample(10) # Gives 10 random rows from the data set
```

```python
df.sample(frac=0.1) # Gives a random 10% of the data from the entire data_set
```
