# Conditional Filtering 

- `Conditional Filtering` allow us to select `Rows` **based on the** `condion on a Columns` 

````python
import numpy as np 
import pandas as pd 
````

````python
df = pd.read_csv('tips.csv')
````

````python
df.head()
````

````python
# Select the data where the total bill is greater then $40
df[df['total_bill'] > 40.00]
````

````python
# Select all the rows where the sex is male 
df[df['sex'] == 'Male']
````

````python
# Select the rows where the size is greater then 3 
df[df['size'] > 3]
````

````python
# Select the rows where the employees are smokers 
df[df['smoker'] == 'Yes']
````

````python
# Select the rows where the waiter are female and smoke 
df[(df['sex'] == 'Female') & (df['smoker'] == 'Yes')]
````

# `.isin()` in Pandas

The `.isin()` method in Pandas is used to filter DataFrame rows or Series elements based on whether they are present in a specified list, set, or other iterable. It is particularly useful for checking membership in a list of values.

---

## **Syntax**
```python
DataFrame.isin(values)


````python
df.head()
````

````python
df[ (df['day'] == 'Sun') | (df['day'] == 'Mon') | (df['day'] == 'Fri') ]
````

````python
weeks = ['Sun', 'Mon', 'Fri']
````

````python
df['day'].isin(weeks)
````

````python
df[df['day'].isin(weeks)]
````
