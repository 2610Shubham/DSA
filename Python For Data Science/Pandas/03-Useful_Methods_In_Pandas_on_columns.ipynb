# **Selecting Columns in a DataFrame**

- **Single Column**:  
  To grab a single column from a DataFrame, use single square brackets `[]`.  
  Example:  
  ```python
  df['column_name']
```

- **Multiple Columns:**
To grab multiple columns, pass a list of column names inside the square brackets [].
Example:
```python
df[['column1', 'column2', 'column3']]
```

# `.apply()` in Pandas

The `.apply()` method in Pandas is used to apply a custom function or a built-in function along an axis (rows or columns) of a DataFrame or Series. It is a versatile and powerful method for performing element-wise operations or aggregations.

---

````python
df['CC Number']
````

`WAP to grabe the last 4 digit of the CC Number`

````python
def last_four(num):
    
    return str(num)[-4:]
````

````python
df['CC Number'].apply(last_four)
````

````python
# Now we will add the new column to our dataset 
df['last_four'] = df['CC Number'].apply(last_four)
````

````python
df.head(1)
````

````python
df['total_bill']
````

````python
def rating_based_on_price(price):
    
    if price < 10:
        return '$'
    elif price >= 10 and price <= 30:
        return '$$'
    else:
        return '$$$'
    
````

````python
df['rating'] = df['total_bill'].apply(rating_based_on_price)
````

# Useful Methods Apply On A Multiple Column

````python
def quality(total_bill, tip):
    
    if tip/total_bill > 0.25:
        return 'Generous'
    else:
        return 'other'
````

````python
df[['total_bill', 'tip']]
````

````python
df[['total_bill', 'tip']].apply( lambda df: quality( df['total_bill'], df['tip']), axis=1) 
````

````python
df['quality'] = df[['total_bill', 'tip']].apply(lambda df: quality(df['total_bill'], df['tip']), axis=1)
````

````python
df.head()
````

````python
df.drop(columns='quality', inplace=True)
````

````python
df.head()
````

````python
df['quality'] = np.vectorize(quality)(df['total_bill'], df['tip'])
````
